# **퍼셉트론이란?**
퍼셉트론은 1957년에 프랑크 로젠블라트(Frank Rosenblatt)가 고안한 알고리즘이며 신경망(딥러닝)의 기원이 되는 알고리즘이다. 
퍼셉트론은 다수의 신호를 입력으로 받아 하나의 신호를 출력하며 퍼셉트론신호는 '흐른다/안 흐른다(1이나 0)'의 두 가지 값을 가질 수 있습니다.

## **퍼셉트론의 기본원리**
(입력신호 x 가중치)들의 총합이 일정한 **임계값**을 넘어가면 1을 출력, 그렇지 않으면 0을 출력한다.

입력신호 또는 출력신호들을 **노드** 또는 **뉴런**이라고 한다.
