# **2.2.1 AND 게이트**
###입력신호가 모두 1일때만 1을출력, 그 외에는 전부 0을 출력
</br>

## <예시>
### 입력 &nbsp; -> &nbsp; 출력  
(0, 0) &nbsp;&nbsp;&nbsp;&nbsp; -> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 0  
(0, 1) &nbsp;&nbsp;&nbsp;&nbsp; -> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 0  
(1, 0) &nbsp;&nbsp;&nbsp;&nbsp; -> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 0  
(1, 1) &nbsp;&nbsp;&nbsp;&nbsp; -> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 1  
</br>
위의 예시를 만족시키는 가중치1, 가중치2, 임계값의 조합설정 예시  
(0.5, 0.5, 0.7)  
(0.5, 0.5, 0.8)  
(1,0, 1.0, 1.0)  


# **2.2.2 NAND 게이트와 OR 게이트**
### **NAND**는 Not AND를 의미하며 그 동작은 AND 게이트의 출력을 뒤집은 것이 된다. 
</br> 
## <예시>
### 입력 &nbsp; -> &nbsp; 출력  
(0, 0) &nbsp;&nbsp;&nbsp;&nbsp; -> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 1  
(0, 1) &nbsp;&nbsp;&nbsp;&nbsp; -> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 1  
(1, 0) &nbsp;&nbsp;&nbsp;&nbsp; -> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 1  
(1, 1) &nbsp;&nbsp;&nbsp;&nbsp; -> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 0  
</br>
위의 예시를 만족시키는 가중치, 임계값 조합은 AND 게이트를 구현하는 매개변수의 부호를 모두 반전시키면 된다.  
ex) (-0.5, -0.5, -0.7)  
</br></br>

### **OR**은 입력신호 중 하나 이상이 1이면 출력이 1이 되는 논리 회리이다.. 
</br> 
## <예시>
### 입력 &nbsp; -> &nbsp; 출력

(0, 1) &nbsp;&nbsp;&nbsp;&nbsp; -> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 1   
(1, 0) &nbsp;&nbsp;&nbsp;&nbsp; -> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 1  
(1, 1) &nbsp;&nbsp;&nbsp;&nbsp; -> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 0  
</br>
위의 예시를 만족시키는 가중치1, 가중치2, 임계값의 조합설정 예시  
(0.5, 0.5, 0.3)  
(1.0, 1.0, 0.5)
